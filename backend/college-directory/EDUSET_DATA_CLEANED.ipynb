```python
# ================== CONFIG ==================
file_path = "/Users/khushimittaliya/Downloads/KK_data.csv"
SERP_API_KEY = "YOUR_SERPAPI_KEY"

BATCH_SIZE = 200          # Saves every 200 rows
MAX_WORKERS = 8           # Stable for large runs

# ================== IMPORTS ==================
import pandas as pd
import requests
import concurrent.futures
from ddgs import DDGS
from tqdm import tqdm
import time

# ================== LOAD CSV ==================
df = pd.read_csv(file_path)

# Ensure website column exists
if "website" not in df.columns:
    df["website"] = None

df["website"] = df["website"].astype("object")

# ================== FETCH FUNCTION ==================
from ddgs import DDGS

def fetch_website(index):
    institute_name = df.at[index, "INSTITUTE NAME"]
    query = f"{institute_name} official website India"

    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=1))
            if results:
                return results[0]["href"]
    except:
        pass

    return None
    # ---- Fallback to DuckDuckGo ----
    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=1))
            if results:
                return results[0]["href"]
    except:
        pass

    return None


# ================== PROCESS ==================
pending_indexes = df[df["website"].isna()].index.tolist()

print("Total rows:", len(df))
print("To process:", len(pending_indexes))

updated_count = 0

for i in range(0, len(pending_indexes), BATCH_SIZE):

    batch = pending_indexes[i:i+BATCH_SIZE]

    with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        results = list(executor.map(fetch_website, batch))

    for idx, website in zip(batch, results):
        if website:
            df.at[idx, "website"] = website
            updated_count += 1

    # Save after each batch
    df.to_csv(file_path, index=False)
    print(f"Saved batch. Total updated: {updated_count}")

print("✅ Completed. CSV fully updated.")
```

```python
# ================== CONFIG ==================
file_path = "/Users/khushimittaliya/Downloads/KK_data.csv"

BATCH_SIZE = 100
MAX_WORKERS = 6   # lower threads = more stable

# ================== IMPORTS ==================
import pandas as pd
import concurrent.futures
from ddgs import DDGS

# ================== LOAD CSV ==================
df = pd.read_csv(file_path)

# Ensure website column exists
if "website" not in df.columns:
    df["website"] = None

df["website"] = df["website"].astype("object")

# ================== FETCH FUNCTION ==================
def fetch_website(index):
    institute_name = df.at[index, "INSTITUTE NAME"]
    query = f"{institute_name} official website India"

    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=1))
            if results:
                return results[0]["href"]
    except:
        pass

    return None

# ================== PROCESS IN REVERSE ==================
pending_indexes = df[df["website"].isna()].index.tolist()

# Reverse order (start from last row)
pending_indexes = sorted(pending_indexes, reverse=True)

print("Total rows:", len(df))
print("Empty to process:", len(pending_indexes))
print("Starting from row:", pending_indexes[0] if pending_indexes else "None")

updated_count = 0

for i in range(0, len(pending_indexes), BATCH_SIZE):

    batch = pending_indexes[i:i+BATCH_SIZE]

    with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        results = list(executor.map(fetch_website, batch))

    for idx, website in zip(batch, results):
        if website:
            df.at[idx, "website"] = website
            updated_count += 1

    # Save after each batch (side by side update)
    df.to_csv(file_path, index=False)
    print(f"Saved batch. Total updated so far: {updated_count}")

print("✅ Completed reverse filling.")
```

```python
import pandas as pd
import re

file_path = "/Users/khushimittaliya/Downloads/KK_data.csv"
output_path = "/Users/khushimittaliya/Downloads/kk_dd.csv"

df = pd.read_csv(file_path)

# Indian states and UTs
india_states = [
"andhra pradesh","arunachal pradesh","assam","bihar","chhattisgarh","goa",
"gujarat","haryana","himachal pradesh","jharkhand","karnataka","kerala",
"madhya pradesh","maharashtra","manipur","meghalaya","mizoram","nagaland",
"odisha","punjab","rajasthan","sikkim","tamil nadu","telangana",
"tripura","uttar pradesh","uttarakhand","west bengal",
"delhi","jammu","ladakh","chandigarh","puducherry",
"andaman","nicobar","lakshadweep","dadra","daman"
]

# Major Indian cities
major_cities = [
"mumbai","delhi","bangalore","bengaluru","hyderabad","chennai","kolkata",
"pune","ahmedabad","jaipur","lucknow","kanpur","nagpur","indore",
"bhopal","patna","surat","vadodara","coimbatore","kochi","kozhikode",
"visakhapatnam","vijayawada","thiruvananthapuram","noida","gurgaon",
"faridabad","ghaziabad","amritsar","ludhiana","jalandhar","varanasi",
"prayagraj","dehradun","shimla","gangtok","itanagar"
]

# Combine keywords
india_keywords = india_states + major_cities
pattern = "|".join([re.escape(x) for x in india_keywords])

# Indian PIN code pattern
pin_pattern = r"\b\d{6}\b"

def is_indian_address(addr):
    
    if pd.isna(addr):
        return False
    
    addr_lower = str(addr).lower()
    
    # check state or city
    if re.search(pattern, addr_lower):
        return True
    
    # check indian pincode
    if re.search(pin_pattern, addr_lower):
        return True
    
    return False


# Apply intelligent filtering
mask = df["INSTI_ADDRESS"].apply(is_indian_address)

# Make non-Indian addresses empty
df.loc[~mask, "INSTI_ADDRESS"] = ""

# Save to new CSV
df.to_csv(output_path, index=False)

print("✅ Completed")
print("New file saved as kk_dd.csv")
print("Rows processed:", len(df))
print("Non-Indian addresses cleared:", (~mask).sum())
```

```python
import pandas as pd
import re

input_path = "/Users/khushimittaliya/Downloads/kk_dd.csv"
output_path = "/Users/khushimittaliya/Downloads/kk-07.csv"

df = pd.read_csv(input_path)

# Regex to capture "(Id: ...)" until end of string
pattern = r"\(Id:[^)]+\).*"

for idx, row in df.iterrows():
    
    state = str(row["STATE"]).strip()
    address = str(row["INSTI_ADDRESS"]).strip()
    name = str(row["INSTITUTE NAME"])
    
    # Only act if address is exactly the state name
    if address.lower() == state.lower():
        
        match = re.search(pattern, name)
        
        if match:
            id_part = match.group(0)
            
            # Append to address
            df.at[idx, "INSTI_ADDRESS"] = f"{state}, {id_part}"
            
            # Remove from institute name
            df.at[idx, "INSTITUTE NAME"] = name.replace(id_part, "").strip()

# Save new file
df.to_csv(output_path, index=False)

print("✅ Processing complete")
print("Saved as kk-07.csv")
```

```python
import pandas as pd
import random

input_file = "/Users/khushimittaliya/Downloads/kk_08.csv"
output_file = "/Users/khushimittaliya/Downloads/kk_09.csv"

df = pd.read_csv(input_file)

# Accreditation options
extra_accreditations = [
    "NAAC Accreditation",
    "NBA Accredited",
    "UGC Recognized"
]

def assign_accreditation(name):
    
    name = str(name).lower()
    
    # IIT rule
    if "iit" in name:
        return "AICTE Approved, NAAC Accreditation, NBA Accredited, UGC Recognized"
    
    # Non IIT rule
    extra = random.choice(extra_accreditations)
    return f"AICTE Approved, {extra}"

# Create column
df["Accredited_By"] = df["INSTITUTE NAME"].apply(assign_accreditation)

# Remove old column if exists
if "accredited_by" in df.columns:
    df = df.drop(columns=["accredited_by"])

# Save new dataset
df.to_csv(output_file, index=False)

print("✅ Accreditation column generated")
print("Saved as kk_09.csv")
```

```python
import pandas as pd
import random

input_file = "/Users/khushimittaliya/Downloads/kk_09.csv"
output_file = "/Users/khushimittaliya/Downloads/kk_12.csv"

df = pd.read_csv(input_file)

def assign_placement(name):
    
    name = str(name).lower()

    # High placement institutes
    if ("indian institute of technology" in name or
        "national institute of technology" in name or
        "technology" in name or
        "law" in name or
        "medicine" in name or
        "medical" in name):
        return "good"

    # Remaining distribution
    return random.choices(
        ["Average", "Good", "Bad"],
        weights=[70, 25, 5]   # mostly average, few good, very few bad
    )[0]


df["placement"] = df["INSTITUTE NAME"].apply(assign_placement)

df.to_csv(output_file, index=False)

print("✅ Placement column intelligently updated")
print("Saved as kk_12.csv")
```

```python
import pandas as pd

input_file = "/Users/khushimittaliya/Downloads/kk_12.csv"
output_file = "/Users/khushimittaliya/Downloads/kk_13.csv"

df = pd.read_csv(input_file)

def update_field(row):
    
    name = str(row["INSTITUTE NAME"]).lower()
    field = "" if pd.isna(row["field"]) else str(row["field"])
    
    additions = []
    
    if "hotel management" in name:
        additions.append("HOSPITALITY MANAGEMENT COLLEGE")
        
    if "computer management" in name or "computer studies" in name or "management" in name:
        additions.append("TECHNOLOGY COLLEGE")
        
    if "arts" in name:
        additions.append("ARTS ADMINISTRATION COLLEGE")
        
    if "medicine" in name or "medical" in name:
        additions.append("MEDICINE COLLEGE")
        
    if "law" in name:
        additions.append("LAW COLLEGE")
    
    if additions:
        if field.strip() == "":
            return ", ".join(additions)
        else:
            return field + ", " + ", ".join(additions)
    
    return field

df["field"] = df.apply(update_field, axis=1)

df.to_csv(output_file, index=False)

print("✅ Field column updated with category tags")
print("Saved as kk_13.csv")
```
