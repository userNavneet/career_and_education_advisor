```python
df = pd.read_csv("student_career_assessment_questions.csv")

df.head()
```

```python
responses = []

print("Career Interest Assessment")
print("--------------------------")

for index, row in df.iterrows():
    
    print("\nQuestion:", row["question"])
    
    print("1 = Strongly Disagree")
    print("2 = Disagree")
    print("3 = Neutral")
    print("4 = Agree")
    print("5 = Strongly Agree")
    
    answer = int(input("Your answer (1-5): "))
    
    responses.append(answer)
```

```python
df["score"] = responses

df.head()
```

```python
category_scores = df.groupby("category")["score"].sum()

category_scores = category_scores.sort_values(ascending=False)

print(category_scores)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=category_scores.values,
    y=category_scores.index
)

plt.title("Career Interest Profile")
plt.xlabel("Interest Score")
plt.ylabel("Career Domain")

plt.show()
```

```python
top_categories = category_scores.head(3)

print("\nTop Career Interests:")
print(top_categories)
```
